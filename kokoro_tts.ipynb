```python
# 1. INSTALL SYSTEM & PYTHON DEPENDENCIES
!apt-get update && apt-get install -y libportaudio2 espeak-ng
!pip install -q kokoro gradio soundfile

# 2. COMPLETE MULTILINGUAL BRANDED GRADIO APPLICATION
import gradio as gr
import numpy as np
from kokoro import KPipeline

pipeline = KPipeline(lang_code='a')

def generate_speech(text, voice, speed):
    global pipeline

    lang_key = voice[0]

    if pipeline.lang_code != lang_key:
        pipeline = KPipeline(lang_code=lang_key)

    generator = pipeline(text, voice=voice, speed=speed)
    all_audio = []

    for _, _, audio in generator:
        all_audio.append(audio)

    if all_audio:
        return (24000, np.concatenate(all_audio))

    return None


choices = {
    "🇺🇸 American (af_/am_)": [
        'af_bella', 'af_sarah', 'af_sky', 'af_nicole', 'af_heart', 'af_alloy',
        'af_aoede', 'af_jessica', 'af_kore', 'af_nova', 'af_river',
        'am_adam', 'am_michael', 'am_echo', 'am_eric', 'am_fenrir',
        'am_liam', 'am_onyx', 'am_puck'
    ],
    "🇬🇧 British (bf_/bm_)": [
        'bf_emma', 'bf_isabella', 'bf_alice', 'bf_lily',
        'bm_george', 'bm_lewis', 'bm_daniel', 'bm_fable'
    ],
    "🇯🇵 Japanese (jf_/jm_)": [
        'jf_alpha', 'jf_gongitsune', 'jf_nezumi', 'jf_tebukuro', 'jm_kumo'
    ],
    "🇨🇳 Mandarin (zf_/zm_)": [
        'zf_xiaobei', 'zf_xiaoni', 'zf_xiaoxiao', 'zf_xiaoyi',
        'zm_yunjian', 'zm_yunxi'
    ],
    "🇪🇸 Spanish (ef_/em_)": [
        'ef_dora', 'em_alex', 'em_santa'
    ],
    "🇫🇷 French (ff_)": [
        'ff_siwis'
    ],
    "🇮🇳 Hindi (hf_/hm_)": [
        'hf_alpha', 'hf_beta', 'hm_omega', 'hm_psi'
    ],
    "🇮🇹 Italian (if_/im_)": [
        'if_sara', 'im_nicola'
    ],
    "🇧🇷 Portuguese (pf_/pm_)": [
        'pf_dora', 'pm_alex'
    ]
}

flat_voices = [v for sublist in choices.values() for v in sublist]

custom_css = """
.gradio-container {
    max-width: 1200px !important;
    margin: auto !important;
}

.vynex-hero {
    border: 2px solid #2563eb;
    border-radius: 16px;
    padding: 24px;
    background: #ffffff;
    text-align: center;
    margin-bottom: 24px;
}

.vynex-hero img {
    width: 100%;
    border-radius: 14px;
    margin-bottom: 22px;
}

.vynex-hero h1 {
    font-size: 36px;
    margin-bottom: 8px;
}

.vynex-hero h3 {
    color: #475569;
    margin-bottom: 12px;
}

.vynex-hero p {
    font-size: 16px;
    max-width: 850px;
    margin: 0 auto 20px auto;
    color: #334155;
}

.vynex-btn {
    display: inline-block;
    padding: 11px 20px;
    margin: 6px;
    border-radius: 8px;
    color: white !important;
    text-decoration: none !important;
    font-weight: 700;
}

.guide-btn {
    background: #2563eb;
}

.youtube-btn {
    background: #ff0000;
}

.website-btn {
    background: #00c853;
}

.vynex-footer {
    text-align: center;
    padding: 20px;
    margin-top: 28px;
    border-top: 1px solid #e5e7eb;
    color: #475569;
}
.facebook-btn {
    background: #1877F2;
}
"""

with gr.Blocks(
    theme=gr.themes.Soft(),
    title="Vynex AI Kokoro TTS",
    css=custom_css
) as demo:

    gr.HTML("""
    <div class="vynex-hero">

        <img src="https://vynex.ai/channel_cover.jpg">

        <h1>🎙️ Vynex AI Kokoro TTS</h1>

        <h3>Free AI Voice Generation Powered by Kokoro TTS</h3>

        <p>
        Generate realistic AI voiceovers for YouTube videos, audiobooks,
        podcasts, narration, educational content, shorts, reels and AI automation workflows.
        </p>

        <a class="vynex-btn guide-btn" href="https://vynex.ai/tools/kokoro-tts" target="_blank">
            📖 Complete Guide
        </a>

        <a class="vynex-btn youtube-btn" href="https://youtube.com/@vynex-ai" target="_blank">
            📺 Subscribe
        </a>

        <a class="vynex-btn website-btn" href="https://vynex.ai" target="_blank">
            🌐 Website
        </a>

        <a class="vynex-btn facebook-btn" href="https://www.facebook.com/profile.php?id=61590397957159" target="_blank">
            👥 Facebook Group
        </a>

    </div>
    """)

    gr.Markdown("## 🎧 Generate AI Voice")

    with gr.Row():
        with gr.Column(scale=2):
            text_input = gr.Textbox(
                label="Input Text",
                value="Hello! Welcome to Vynex AI Kokoro TTS. Generate realistic AI voices for YouTube videos, narration, podcasts, and content creation.",
                lines=6
            )

            with gr.Row():
                voice_input = gr.Dropdown(
                    choices=flat_voices,
                    value='am_adam',
                    label="Voice Model"
                )

                speed_input = gr.Slider(
                    minimum=0.5,
                    maximum=2.0,
                    value=1.0,
                    step=0.1,
                    label="Speed"
                )

            run_btn = gr.Button("🎙️ Generate Speech", variant="primary")

        with gr.Column(scale=1):
            audio_output = gr.Audio(
                label="Generated Audio",
                type="numpy"
            )

            gr.Markdown("""
### 🌍 Voice Legend

- **af/am**: US English
- **bf/bm**: UK English
- **jf/jm**: Japanese
- **zf/zm**: Mandarin
- **ef/em**: Spanish
- **ff**: French
- **hf/hm**: Hindi
- **if/im**: Italian
- **pf/pm**: Portuguese
            """)

    run_btn.click(
        fn=generate_speech,
        inputs=[text_input, voice_input, speed_input],
        outputs=audio_output
    )

    gr.HTML("""
    <div class="vynex-footer">
        <b>Made with ❤️ by Vynex AI</b><br>
        AI Tools • YouTube Automation • AI Voice Generation<br><br>
        <a href="https://vynex.ai" target="_blank">Website</a> |
        <a href="https://youtube.com/@vynex-ai" target="_blank">YouTube</a> |
        <a href="https://vynex.ai/tools/kokoro-tts" target="_blank">Complete Guide</a>
    </div>
    """)

demo.launch(share=True, debug=True)
```

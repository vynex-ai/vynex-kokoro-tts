```python
# 1. INSTALL SYSTEM & PYTHON DEPENDENCIES
# espeak-ng is the phonetic engine; libportaudio2 is the audio driver.
!apt-get update && apt-get install -y libportaudio2 espeak-ng
!pip install -q kokoro gradio soundfile

# 2. COMPLETE MULTILINGUAL GRADIO APPLICATION
import gradio as gr
import numpy as np
import soundfile as sf
from kokoro import KPipeline

# Initialize the primary US English pipeline
# The app will dynamically switch this when you change languages
pipeline = KPipeline(lang_code='a')

def generate_speech(text, voice, speed):
    global pipeline

    # Logic to switch language based on voice prefix
    # Voice pattern: af_ (American), bf_ (British), jf_ (Japanese), etc.
    lang_key = voice[0] # 'a', 'b', 'j', 'z', 'e', 'f', 'h', 'i', 'p'

    if pipeline.lang_code != lang_key:
        pipeline = KPipeline(lang_code=lang_key)

    # Generate audio
    generator = pipeline(text, voice=voice, speed=speed)
    all_audio = []

    for _, _, audio in generator:
        all_audio.append(audio)

    if all_audio:
        return (24000, np.concatenate(all_audio))
    return None

# FULL VOICE DIRECTORY
choices = {
    "рџ‡єрџ‡ё American (af_/am_)": [
        'af_bella', 'af_sarah', 'af_sky', 'af_nicole', 'af_heart', 'af_alloy',
        'af_aoede', 'af_jessica', 'af_kore', 'af_nova', 'af_river',
        'am_adam', 'am_michael', 'am_echo', 'am_eric', 'am_fenrir', 'am_liam', 'am_onyx', 'am_puck'
    ],
    "рџ‡¬рџ‡§ British (bf_/bm_)": [
        'bf_emma', 'bf_isabella', 'bf_alice', 'bf_lily',
        'bm_george', 'bm_lewis', 'bm_daniel', 'bm_fable'
    ],
    "рџ‡Їрџ‡µ Japanese (jf_/jm_)": ['jf_alpha', 'jf_gongitsune', 'jf_nezumi', 'jf_tebukuro', 'jm_kumo'],
    "рџ‡Ёрџ‡і Mandarin (zf_/zm_)": ['zf_xiaobei', 'zf_xiaoni', 'zf_xiaoxiao', 'zf_xiaoyi', 'zm_yunjian', 'zm_yunxi'],
    "рџ‡Єрџ‡ё Spanish (ef_/em_)": ['ef_dora', 'em_alex', 'em_santa'],
    "рџ‡«рџ‡· French (ff_)": ['ff_siwis'],
    "рџ‡®рџ‡і Hindi (hf_/hm_)": ['hf_alpha', 'hf_beta', 'hm_omega', 'hm_psi'],
    "рџ‡®рџ‡№ Italian (if_/im_)": ['if_sara', 'im_nicola'],
    "рџ‡§рџ‡· Portuguese (pf_/pm_)": ['pf_dora', 'pm_alex']
}

# Flatten the choices for the dropdown
flat_voices = [v for sublist in choices.values() for v in sublist]

# UI DESIGN
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# рџЊё Kokoro Universal TTS Studio (2026)")
    gr.Markdown("Supports 8 languages and 50+ voices. Just select a voice and type!")

    with gr.Row():
        with gr.Column(scale=2):
            text_input = gr.Textbox(
                label="Input Text",
                value="Hello! I am Kokoro, an 82M parameter TTS model. I'm incredibly fast and natural.",
                lines=6
            )
            with gr.Row():
                voice_input = gr.Dropdown(choices=flat_voices, value='am_adam', label="Voice Model")
                speed_input = gr.Slider(0.5, 2.0, 1.0, step=0.1, label="Speed")

            run_btn = gr.Button("Generate Speech", variant="primary")

        with gr.Column(scale=1):
            audio_output = gr.Audio(label="Generated Audio", type="numpy")
            gr.Markdown("### Voice Legend\n- **af/am**: US English\n- **bf/bm**: UK English\n- **jf/jm**: Japanese\n- **zf/zm**: Mandarin")

    run_btn.click(generate_speech, [text_input, voice_input, speed_input], audio_output)

demo.launch(share=True, debug=True)
```
